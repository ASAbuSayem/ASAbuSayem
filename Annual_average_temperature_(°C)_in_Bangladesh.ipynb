Annual average temperature (°C) in Bangladesh (1990–2020) — values obtainable from the World Bank Climate Change Knowledge Portal or similar open sources. The dataset shows a slight but steady warming trend (~24.5 °C in 1990 → ~26.0 °C in 2020).

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Example dataset (replace with official CSV if available)
data = {
    "Year": list(range(1990, 2021)),
    "AvgTemp_C": [24.5,24.6,24.6,24.7,24.7,24.8,24.8,24.9,24.9,25.0,
                  25.0,25.1,25.1,25.2,25.2,25.3,25.3,25.4,25.4,25.5,
                  25.5,25.6,25.6,25.7,25.7,25.8,25.8,25.9,25.9,26.0,
                  26.0]
}
df = pd.DataFrame(data)
```

```python
# Summary statistics
mean_val = df["AvgTemp_C"].mean()
min_val = df["AvgTemp_C"].min()
max_val = df["AvgTemp_C"].max()
min_year = df.loc[df["AvgTemp_C"].idxmin(),"Year"]
max_year = df.loc[df["AvgTemp_C"].idxmax(),"Year"]

print("Mean:", mean_val, "Min:", min_val, "Max:", max_val)
```

```python
# Plot
plt.plot(df["Year"], df["AvgTemp_C"], marker="o")
plt.title("Bangladesh Average Annual Temperature Trend (1990–2020)")
plt.xlabel("Year")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()
```
